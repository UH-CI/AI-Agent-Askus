# Question Answerer Example

This notebook demonstrates how to use the `QuestionAnswerer` class to extract answers from Wikipedia data.

```python
import jsonlines
import json
import numpy as np
import base64

# Import our QuestionAnswerer class
from wiki_question_answerer import QuestionAnswerer, clean_wikipedia_html
```

## Load Sample Data

We'll load data from the Natural Questions dataset.

```python
# Path to your jsonlines file
jsonl_path = 'v1.0-simplified_nq-dev-all.jsonl'

# Load a sample record
sample_records = []
with jsonlines.open(jsonl_path) as reader:
    for idx, obj in enumerate(reader):
        if idx < 5:  # Get the first 5 examples
            sample_records.append(obj)
        else:
            break

print(f"Loaded {len(sample_records)} sample records")
```

## Examine the First Record

Let's look at the structure of the first record.

```python
# Check what keys are available in the record
print("Available keys in the first record:")
print(sample_records[0].keys())

# Print the question text
print("\nQuestion:")
print(sample_records[0]['question_text'])
```

## Clean and Display Document Content

Let's see the cleaned content from the Wikipedia article.

```python
# Get a sample of the cleaned HTML content
clean_text = clean_wikipedia_html(sample_records[0]['document_html'])

# Display the first 500 characters
print(clean_text[:500] + "...")
```

## Answer a Question

Now let's use our QuestionAnswerer class to answer the question from the first record.

```python
# Create a QuestionAnswerer for the first record
qa = QuestionAnswerer(sample_records[0])

# Get the answer using the question in the record
answer_result = qa.answer_question()

# Display the results
print(f"Question: {answer_result['question']}")
print(f"Answer: {answer_result['answer_text']}")
print(f"Answer type: {answer_result['answer_type']}")
print(f"Candidate index: {answer_result['candidate_index']}")
print("\nContext:")
print(answer_result['context'][:300] + "..." if len(answer_result['context']) > 300 else answer_result['context'])
```

## Try with Different Records

Let's try answering questions from the other records we loaded.

```python
# Process all loaded records
for idx, record in enumerate(sample_records):
    print(f"\n===== Record {idx} =====")
    
    # Create a QuestionAnswerer
    qa = QuestionAnswerer(record)
    
    # Get the answer
    answer_result = qa.answer_question()
    
    # Display the results
    print(f"Question: {answer_result['question']}")
    print(f"Answer: {answer_result['answer_text']}")
    print(f"Answer type: {answer_result['answer_type']}")
```

## Custom Question

We can also pass a custom question to the answer_question method, which might be useful when you want to try different phrasings.

```python
# Create a QuestionAnswerer for the first record
qa = QuestionAnswerer(sample_records[0])

# Try a custom question (rephrasing of the original)
custom_question = "What is the main topic of this Wikipedia article?"
answer_result = qa.answer_question(custom_question)

# Display the results
print(f"Custom question: {custom_question}")
print(f"Answer: {answer_result['answer_text']}")
print(f"Answer type: {answer_result['answer_type']}")
print(f"Context: {answer_result['context'][:200]}..." if len(answer_result['context']) > 200 else answer_result['context'])
```
