## Import and Use the Question Generator

Now we can import the `QuestionGenerator` class from our script and use it directly:

```python
from generate_policy_questions import QuestionGenerator

# Initialize the question generator
api_key = os.environ.get("GOOGLE_API_KEY")
generator = QuestionGenerator(api_key=api_key)
```

## Generate Questions for a Single Document

Let's try generating questions for a single document first:

```python
import json

# Load the policies
with open('testing/policies.json', 'r') as f:
    policies = json.load(f)

# Generate questions for the first policy
first_policy = policies[0]
questions = generator.generate_questions(first_policy, num_questions=3)

# Display the questions
for q in questions:
    print(f"ID: {q['id']}")
    print(f"Question: {q['question']}")
    print(f"Document ID: {q['document_id']}")
    print()
```

## Generate Questions for All Documents

Now let's generate questions for all documents in the policies.json file:

```python
# Process a small sample of documents (e.g., the first 2)
sample_policies = policies[:2]
all_questions = generator.process_documents(sample_policies, num_questions=3)

# Print the number of questions generated
print(f"Generated {len(all_questions)} questions from {len(sample_policies)} documents.")
```

## Save the Questions to a File

We can save the generated questions to a JSON file:

```python
# Create the data directory if it doesn't exist
import os
os.makedirs('data', exist_ok=True)

# Save the questions
output_path = 'data/sample_evaluation_questions.json'
generator.save_questions(all_questions, output_path)
print(f"Saved questions to {output_path}")
```
